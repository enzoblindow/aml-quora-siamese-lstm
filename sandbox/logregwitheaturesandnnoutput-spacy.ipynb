```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
import statsmodels.api as sm
```

```python
nn_out = pd.read_csv('../spacy_lstm_train_weights.csv')
```

```python
nn_out.columns = ['nn_out', 'id']
nn_out.head()
```

```python
train = pd.read_csv('../data/train_with_sim_and_ents_long.csv')
train.head()
```

```python
merged = train.merge(nn_out)
```

```python
merged['last_char'] = np.where(merged.question1.str[-1:] == merged.question2.str[-1:], 1, 0)
merged.head()
```

```python
y = merged['is_duplicate']
x = merged[['similarity_score', 'nn_out', 'last_char']]
x.head(20)
```

```python
from scipy import stats
stats.chisqprob = lambda chisq, df: stats.chi2.sf(chisq, df)
```

```python
logit_model=sm.Logit(y,x)
result=logit_model.fit()
print(result.summary())
```

```python
nn_out_test = pd.read_csv('../spacy_LSTM_5epochs.csv')
nn_out_test.columns = ['nn_out','test_id']
nn_out_test.head()
```

```python
test = pd.read_csv('../data/test_with_sim_and_ents_long.csv')
test_merged = test.merge(nn_out_test)
test_merged['last_char'] = np.where(test_merged.question1.str[-1:] == test_merged.question2.str[-1:], 1, 0)
test_merged.head()
```

```python
x_test = test_merged[['similarity_score', 'nn_out', 'last_char']]
```

```python
logreg = LogisticRegression()
logreg.fit(x, y)
y_pred = logreg.predict(x_test)
```

```python
len(y_pred)
```

```python
predict = pd.DataFrame({"test_id":test_merged['test_id'], "is_duplicate":y_pred})
predict.to_csv("logregattempt2.csv", index=False)
```
